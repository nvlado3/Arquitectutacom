# Primera entrega<a name="id1"></a>
Números primos: son aquellos números que únicamente pueden ser divisibles por si mismos y por uno (1); los números pares al ser divisibles por el dos (2) no son primos la única excepción a esta regla es el mismo numero dos (2).
## Abstraer el problema
Se debe diseñar o crear un algoritmo que permita la entrada de un numero entero; y este a su vez debe validar si corresponde o no a un numero primo.
## Estructurar en un algoritmo.
* Pedir al usuario que introduzca un numero mayor que cero.
* Valida que el numero introducido sea mayor que cero.
* Si es menor o igual que cero; le informa al usuario y finaliza.
* En caso contrario, y con el animo de evitar muchas interacciones revisa si el numero dado es par y diferente de dos (2); en caso que así sea significa que no es primo.
* Luego comienza a validar el residuo o modulo de la división entre el numero dado y cada numero desde uno hasta el numero dado.
* Si en algún momento el residuo llegar a ser cero quiere decir que el numero no es primo porqué es divisible. En el resto de opciones el numero es primo.

    public static boolean esPar(int numero) {

        boolean par;
        if ((numero % 2) == 0) {
            par = true;
        } else {
            par = false;
        }
        return par;
    }

    public static boolean esPrimo(int numero) {

        boolean primo = true;
        int x = 0;
        if (esPar(numero) && numero != 2) {
            primo = false;
        } else {
            for (int i = 1; i <= numero; i++) {
                if (numero % i == 0) {
                    x++;
                }
                if (x == 2) {
                    primo = true;
                } else if (x >= 3) {
                    primo = false;
                    break;
                } else {
                    primo = false;
                }
            }
        }
        return primo;
    }

## 3. Descripción breve de los modos de direccionamiento que serán posibles.
### Direccionamiento Directo 
El modo de direccionamiento inmediato se refiere a aquellas instrucciones en los que uno de los operandos es un valor constante que viene dentro de la instrucción.

Por ejemplo:

addi \\$s6, \\$s7, 8

Corresponde a \\$s6 = \\$s7 +8 donde 8 sería direccionado de forma inmediata por ser una constante que haría parte de la instrucción

### Directo a registro: 

En este modo de direccionamiento se incluye en la instrucción la dirección del registro que se desea acceder, como fuente o como destino.

Por ejemplo:

add \\$s1, \\$s2, \\$s3

Equivale a \\$s1=\\$s2+\\$s3 de manera que se envían como parámetros de la instrucción las direcciones de los registros \\$s1 (destino), \\$s2 (fuente) y \\$s3 (fuente)

## CREACION DE REGISTROS  
Para saber si un numero es primo o no, se requiere un numero inicial el cual deberá guardarse (registro1); así mismo se comenzará a dividir este número iniciando en uno incrementándose en cero, al cual llamaré contador y es necesario guardarlo (registro2), una vez se comience a dividir se comparará el residuo del número, en caso de ser cero significa que el número es divisible por ese número, técnicamente un numero primo solo debe ser divisible dos veces sin que quede residuo, cuando se divide por el 1 y por si mismo, por lo cual es necesario que cada vez que el número inicial sea divisible por algún número, es decir que no tenga residuo, se guarde de manera temporal este contador auxiliar (registro 3) para luego comparar y determinar si es o no primo. Adicionalmente se debe tener en cuenta el registro cero que es del sistema.

Así las cosas, únicamente necesitaremos 4 registros.
Registro 0 = reservado para el sistema 
Registro 1 = número inicial (tipo numérico)
Registro 2 = contador de interacciones. (tipo numérico)
Registro 3 = numero de veces que el valor inicial es divisible sin residuo (tipo numérico)

No obstante se van a crear cuatro regitros adicionales en caso para tener 2 elevado a la 3 posiciones de almacenamiento.
